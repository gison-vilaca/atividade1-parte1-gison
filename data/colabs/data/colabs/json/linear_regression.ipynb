```python
df2.columns = df2.columns.str.replace('-', '_')

df2_numerico = df2.select_dtypes(include=np.number)

# 3. Calcula a correlação com a variável alvo 'pontos'
print("="*60)
print("CORRELAÇÃO DAS VARIÁVEIS COM 'PONTOS' DO JOGADOR (df2)")
print("="*60)

# Calcula a matriz de correlação e isola a coluna 'pontos'
correlacoes_pontos_jogador = df2_numerico.corr()['pontos'].sort_values(ascending=False)

# Remove 'pontos' de si mesmo e exibe as 10 mais fortes (excluindo aquelas que já são derivadas, como 'eficiencia_simples')
variaveis_fortes_pontos = correlacoes_pontos_jogador.drop(
    ['pontos', 'eficiencia_simples'], 
    errors='ignore'
).head(10)

print(variaveis_fortes_pontos)
```

```python
formula_pontos_jogador = (
    "pontos ~ minutos + lances_livres_convertidos + assistencias + erros"
)

# Ajustando o modelo no df2 (Dados do Jogador)
modelo_pontos_jogador = smf.ols(formula_pontos_jogador, data=df2).fit()

# Apresentando os resultados
print("="*80)
print("⛹️ RESULTADOS DA REGRESSÃO LINEAR MÚLTIPLA (Pontos do Jogador)")
print("="*80)
print(modelo_pontos_jogador.summary())
```

```python
correlacoes_rebotes_jogador = df2_numerico.corr()['rebotes_totais'].sort_values(ascending=False)
```

```python
formula_rebotes_jogador = (
    "rebotes_totais ~ rebotes_defensivos + tocos + altura_cm + faltas"
)

modelo_rebotes_jogador = smf.ols(formula_rebotes_jogador, data=df2).fit()

# Apresentando os resultados
print("="*80)
print("⛹️ RESULTADOS DA REGRESSÃO LINEAR MÚLTIPLA (Rebotes do Jogador)")
print("="*80)
print(modelo_rebotes_jogador.summary())
```
